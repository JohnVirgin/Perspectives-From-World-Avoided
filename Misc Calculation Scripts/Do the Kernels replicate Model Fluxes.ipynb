```python
Source_File = nc.Dataset('Kernels/Interpolated/CAM3_Kernels.nc')
#CAM3 Horizontal resolution grid
Lat = np.squeeze(Source_File.variables['lat'])
Lon = np.squeeze(Source_File.variables['lon'])
```

```python
#net radiation first

RCP45_LW = np.mean(np.swapaxes(pk.load(open(\
    'Future Projection CC Responses/RCP45_dFLUT_CAM3Grid.pickle',\
    'rb'),encoding='latin1'),2,3),axis=1)
RCP45_SW = np.mean(np.swapaxes(pk.load(open(\
    'Future Projection CC Responses/RCP45_dFSNT_CAM3Grid.pickle',\
    'rb'),encoding='latin1'),2,3),axis=1)
wAVD_LW = np.mean(np.swapaxes(pk.load(open(\
    'Future Projection CC Responses/wAVD_dFLUT_CAM3Grid.pickle',\
    'rb'),encoding='latin1'),2,3),axis=1)
wAVD_SW = np.mean(np.swapaxes(pk.load(open(\
    'Future Projection CC Responses/wAVD_dFSNT_CAM3Grid.pickle',\
    'rb'),encoding='latin1'),2,3),axis=1)
```

```python
RCP45_NET = RCP45_SW-RCP45_LW
wAVD_NET = wAVD_SW-wAVD_LW
```

```python
RCP45_NET_AA = np.mean(aa.LatLonavg_Time(RCP45_NET[:,54:,:],Lat[54:],Lon))
RCP45_NET_GA = np.mean(aa.LatLonavg_Time(RCP45_NET[:,:,:],Lat[:],Lon))
wAVD_NET_AA = np.mean(aa.LatLonavg_Time(wAVD_NET[:,54:,:],Lat[54:],Lon))
wAVD_NET_GA = np.mean(aa.LatLonavg_Time(wAVD_NET[:,:,:],Lat[:],Lon))
```
