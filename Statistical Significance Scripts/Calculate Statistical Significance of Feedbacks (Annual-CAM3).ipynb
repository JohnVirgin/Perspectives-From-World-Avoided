```python
#CAM3 Horizontal resolution grid
LLL3 = Dataset('Kernels/Interpolated/CAM3_Kernels.nc')
CAM3_Lat = np.squeeze(LLL3.variables['lat'])
CAM3_Lon = np.squeeze(LLL3.variables['lon'])

#Lists for iterating
Months = np.arange(0,12)
FP_Ensemble_Members = np.arange(0,5)

RCP45_dST_CAM3Grid = np.swapaxes(pickle.load(open(\
   "Future Projection CC Responses/RCP45_dST_CAM3Grid.pickle","rb"),encoding='latin1'),2,3)

wAVD_dST_CAM3Grid = np.swapaxes(pickle.load(open(\
   "Future Projection CC Responses/wAVD_dST_CAM3Grid.pickle","rb"),encoding='latin1'),2,3)

RCP45_dST_CAM3Grid_AA = np.zeros([5,12])
RCP45_dST_CAM3Grid_GA = np.zeros([5,12])

wAVD_dST_CAM3Grid_AA = np.zeros([5,12])
wAVD_dST_CAM3Grid_GA = np.zeros([5,12])

for i in range(FP_Ensemble_Members.size):
    RCP45_dST_CAM3Grid_AA[i,:] = Area_Avg.LatLonavg_Time(\
    RCP45_dST_CAM3Grid[i,:,54:,:],CAM3_Lat[54:],CAM3_Lon)
    
    RCP45_dST_CAM3Grid_GA[i,:] = Area_Avg.LatLonavg_Time(\
    RCP45_dST_CAM3Grid[i,:,:,:],CAM3_Lat,CAM3_Lon)
    
    wAVD_dST_CAM3Grid_AA[i,:] = Area_Avg.LatLonavg_Time(\
    wAVD_dST_CAM3Grid[i,:,54:,:],CAM3_Lat[54:],CAM3_Lon)
    
    wAVD_dST_CAM3Grid_GA[i,:] = Area_Avg.LatLonavg_Time(\
    wAVD_dST_CAM3Grid[i,:,:,:],CAM3_Lat,CAM3_Lon)
    
RCP45_dST_CAM3Grid_AA = np.mean(RCP45_dST_CAM3Grid_AA,axis=1)
RCP45_dST_CAM3Grid_GA = np.mean(RCP45_dST_CAM3Grid_GA,axis=1)

wAVD_dST_CAM3Grid_AA = np.mean(wAVD_dST_CAM3Grid_AA,axis=1)
wAVD_dST_CAM3Grid_GA = np.mean(wAVD_dST_CAM3Grid_GA,axis=1)

RCP45_OCN_Feedback_AA = RCP45_OCN_EnergyB_AA/RCP45_dST_CAM3Grid_AA
wAVD_OCN_Feedback_AA = wAVD_OCN_EnergyB_AA/wAVD_dST_CAM3Grid_AA

RCP45_OCN_Feedback_GA = RCP45_OCN_EnergyB_GA/RCP45_dST_CAM3Grid_GA
wAVD_OCN_Feedback_GA = wAVD_OCN_EnergyB_GA/wAVD_dST_CAM3Grid_GA
```

## Calculate Statistical Significance using a Mann-Whitney U test

```python
StatSig_LR_EnergyB_AA = stats.mannwhitneyu(\
    CAM3_RCP45_LR_EnergyB_AA,CAM3_wAVD_LR_EnergyB_AA,alternative = 'two-sided')
StatSig_ALB_EnergyB_AA = stats.mannwhitneyu(\
    CAM3_RCP45_ALB_EnergyB_AA,CAM3_wAVD_ALB_EnergyB_AA,alternative = 'two-sided')
StatSig_WV_EnergyB_AA = stats.mannwhitneyu(\
    CAM3_RCP45_WV_EnergyB_AA,CAM3_wAVD_WV_EnergyB_AA,alternative = 'two-sided')
StatSig_CLD_LW_EnergyB_AA = stats.mannwhitneyu(\
    CAM3_RCP45_CLD_LW_EnergyB_AA,CAM3_wAVD_CLD_LW_EnergyB_AA,alternative = 'two-sided')

StatSig_CLD_SW_EnergyB_AA = stats.mannwhitneyu(\
    CAM3_wAVD_CLD_SW_EnergyB_AA,CAM3_RCP45_CLD_SW_EnergyB_AA,alternative = 'two-sided')

StatSig_DYNatmos_EnergyB_AA = stats.mannwhitneyu(\
    RCP45_DYNatmos_EnergyB_AA,wAVD_DYNatmos_EnergyB_AA,alternative = 'two-sided')
StatSig_OCN_EnergyB_AA = stats.mannwhitneyu(\
    RCP45_OCN_EnergyB_AA,wAVD_OCN_EnergyB_AA,alternative = 'two-sided')

StatSig_LR_EnergyB_GA = stats.mannwhitneyu(\
    CAM3_wAVD_LR_EnergyB_GA,CAM3_RCP45_LR_EnergyB_GA,alternative = 'two-sided')
StatSig_CLD_LW_EnergyB_GA = stats.mannwhitneyu(\
    CAM3_wAVD_CLD_LW_EnergyB_GA,CAM3_RCP45_CLD_LW_EnergyB_GA,alternative = 'two-sided')

StatSig_ALB_EnergyB_GA = stats.mannwhitneyu(\
    CAM3_RCP45_ALB_EnergyB_GA,CAM3_wAVD_ALB_EnergyB_GA,alternative = 'two-sided')
StatSig_WV_EnergyB_GA = stats.mannwhitneyu(\
    CAM3_RCP45_WV_EnergyB_GA,CAM3_wAVD_WV_EnergyB_GA,alternative = 'two-sided')
StatSig_CLD_SW_EnergyB_GA = stats.mannwhitneyu(\
    CAM3_RCP45_CLD_SW_EnergyB_GA,CAM3_wAVD_CLD_SW_EnergyB_GA,alternative = 'two-sided')
StatSig_OCN_EnergyB_GA = stats.mannwhitneyu(\
    RCP45_OCN_EnergyB_GA,wAVD_OCN_EnergyB_GA,alternative = 'two-sided')
```

```python
StatSig_LR_Feedback_AA = stats.mannwhitneyu(\
    CAM3_RCP45_LR_Feedback_AA,CAM3_wAVD_LR_Feedback_AA,alternative = 'two-sided')
StatSig_ALB_Feedback_AA = stats.mannwhitneyu(\
    CAM3_RCP45_ALB_Feedback_AA,CAM3_wAVD_ALB_Feedback_AA,alternative = 'two-sided')
StatSig_WV_Feedback_AA = stats.mannwhitneyu(\
    CAM3_RCP45_WV_Feedback_AA,CAM3_wAVD_WV_Feedback_AA,alternative = 'two-sided')
StatSig_CLD_LW_Feedback_AA = stats.mannwhitneyu(\
    CAM3_RCP45_CLD_LW_Feedback_AA,CAM3_wAVD_CLD_LW_Feedback_AA,alternative = 'two-sided')

StatSig_CLD_SW_Feedback_AA = stats.mannwhitneyu(\
    CAM3_wAVD_CLD_SW_Feedback_AA,CAM3_RCP45_CLD_SW_Feedback_AA,alternative = 'two-sided')

StatSig_DYNatmos_Feedback_AA = stats.mannwhitneyu(\
    RCP45_DYNatmos_Feedback_AA,wAVD_DYNatmos_Feedback_AA,alternative = 'two-sided')
StatSig_OCN_Feedback_AA = stats.mannwhitneyu(\
    RCP45_OCN_Feedback_AA,wAVD_OCN_Feedback_AA,alternative = 'two-sided')

StatSig_LR_Feedback_GA = stats.mannwhitneyu(\
    CAM3_wAVD_LR_Feedback_GA,CAM3_RCP45_LR_Feedback_GA,alternative = 'two-sided')
StatSig_CLD_LW_Feedback_GA = stats.mannwhitneyu(\
    CAM3_wAVD_CLD_LW_Feedback_GA,CAM3_RCP45_CLD_LW_Feedback_GA,alternative = 'two-sided')

StatSig_ALB_Feedback_GA = stats.mannwhitneyu(\
    CAM3_RCP45_ALB_Feedback_GA,CAM3_wAVD_ALB_Feedback_GA,alternative = 'two-sided')
StatSig_WV_Feedback_GA = stats.mannwhitneyu(\
    CAM3_RCP45_WV_Feedback_GA,CAM3_wAVD_WV_Feedback_GA,alternative = 'two-sided')
StatSig_CLD_SW_Feedback_GA = stats.mannwhitneyu(\
    CAM3_RCP45_CLD_SW_Feedback_GA,CAM3_wAVD_CLD_SW_Feedback_GA,alternative = 'two-sided')
StatSig_OCN_Feedback_GA = stats.mannwhitneyu(\
    RCP45_OCN_Feedback_GA,wAVD_OCN_Feedback_GA,alternative = 'two-sided')
```

```python
#The critical value of U at 0.05 is 2 in this case (n1,n2=5)
#we reject the null hypothesis if U is equal to or less than 2
#the null hypothesis is that these populations are equal. 

#i.e. if U is greater than 2, the populations are not statistically significantly different
```
